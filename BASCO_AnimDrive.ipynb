# **CONVERTING THE ANIME TO MP4 (A SUPPORTED FORMAT BY MOST PLATFORMS)**
For converting the anime we need just to edit the anime path my case ex:"/content/drive/MyDrive/anime/One Piece/" the script will convert any video inside the folder to the MP4 format and delete the original files to keep more space in drive and not make it out of quota .

```python
!cd "/content/drive/MyDrive/anime/Tokyo Ghoul/" && mkdir "temp" && for i in *.*; do mv $i temp && ffmpeg -y -vsync 0 -hwaccel cuda -hwaccel_output_format cuda -i temp/$i -c:v h264_nvenc -preset fast -rc ll_2pass_size -b:v 1M -c:a copy ./${i%.*}.mp4 && rm -rf temp/$i; done; rm -rf "temp"
```
